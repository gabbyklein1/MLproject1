```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

data = pd.read_csv('./merge_data_update.csv', index_col=0)
```

# Housing Features

#### GrLivArea: Above grade (ground) living area square feet

```python
print('Column Name: ', jude_cols[1])
print('Number of Null Values: ', np.sum(data['GrLivArea'].isnull()))
print("Max: ", data['GrLivArea'].max())
print("Min: ", data['GrLivArea'].min())
print(data['GrLivArea'].value_counts())
data['GrLivArea'].hist(weights=np.ones(len(data['GrLivArea']))/len(data['GrLivArea']))
plt.show()
plt.scatter(x=data['GrLivArea'], y=data['SalePrice'], alpha=.2)
plt.show()
```
